```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/4GeeksAcademy/naive-bayes-project-tutorial/main/playstore_reviews.csv"
```

```python
df= pd.read_csv(url)
```

```python
df.shape
```

El dataFrame se conforma de 891 registros y 3 variables.

```python
df.info()
```

El df contiene 2 variables categoricas y una variable numerica.

```python
df.head()
```

```python
from numpy._core.defchararray import lower
df.drop(columns = "package_name")
df["review"] =  df["review"].str.strip().str.lower()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X = df["review"]
y = df["polarity"]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1234)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

vec_model = CountVectorizer()
X_train_vec = vec_model.fit_transform(X_train)
X_test_vec = vec_model.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

model_mnb = MultinomialNB()
model_mnb.fit(X_train_vec, y_train)
y_pred_mnb = model_mnb.predict(X_test_vec)
print(classification_report(y_pred_mnb, y_test))

model_bnb = BernoulliNB()
model_bnb.fit(X_train_vec, y_train)
y_pred_bnb = model_bnb.predict(X_test_vec)
print(classification_report(y_pred_bnb, y_test))
```

Conclusión:
El modelo Multinomial Naive Bayes presentó un mejor desempeño que Bernoulli Naive Bayes, alcanzando mayor accuracy y un F1-score superior, especialmente en la clase minoritaria. Además, mostró un alto recall para esta clase, lo que indica una buena capacidad para identificar correctamente las reseñas positivas/negativas, aunque con cierta pérdida de precisión. En conjunto, Multinomial Naive Bayes resulta más adecuado para la clasificación de reseñas de Google Play, ya que aprovecha mejor la frecuencia de las palabras en el texto.
